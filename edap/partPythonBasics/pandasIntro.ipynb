# Pandas Introduction

[pandas](https://pandas.pydata.org/) is the standard Python library for working with tabular (structured) data. It provides fast, flexible data structures — primarily **Series** and **DataFrame** — that make data manipulation, cleaning, and analysis straightforward.

```{tip}
If you have ever worked with spreadsheets (Excel, Google Sheets), think of a DataFrame as a programmable spreadsheet.
```

````python
import pandas as pd
````

## Series vs DataFrame

A **Series** is a one-dimensional labeled array — essentially a single column of data with an index.

A **DataFrame** is a two-dimensional labeled data structure — a table with rows and columns, where each column is a Series.

````python
# A Series
temperatures = pd.Series([22.1, 23.4, 21.8, 24.0], name="Temperature")
temperatures
````

````python
type(temperatures)
````

## Creating DataFrames

DataFrames can be created from Python dictionaries, lists of lists, or many other sources.

### From a dictionary

````python
data = {
    "room": ["Living Room", "Bedroom", "Kitchen", "Office"],
    "temperature": [22.1, 20.5, 23.4, 21.8],
    "humidity": [45.0, 50.2, 38.7, 42.1]
}

df = pd.DataFrame(data)
df
````

### From a list of lists

````python
rows = [
    ["Living Room", 22.1, 45.0],
    ["Bedroom", 20.5, 50.2],
    ["Kitchen", 23.4, 38.7],
    ["Office", 21.8, 42.1]
]

df2 = pd.DataFrame(rows, columns=["room", "temperature", "humidity"])
df2
````

## Basic Indexing

There are several ways to select data from a DataFrame.

### Select a single column (returns a Series)

````python
df["temperature"]
````

```{note}
Selecting a single column with `df["col"]` returns a **Series**.
```

### Select multiple columns (returns a DataFrame)

````python
df[["room", "temperature"]]
````

```{note}
Selecting multiple columns with `df[["col1", "col2"]]` returns a **DataFrame**.
```

### Label-based indexing with `.loc[]`

`.loc[]` selects rows and columns by **label** (index value and column name).

````python
# Select row with index label 0
df.loc[0]
````

````python
# Select rows 0 to 2, columns "room" and "temperature"
df.loc[0:2, ["room", "temperature"]]
````

### Position-based indexing with `.iloc[]`

`.iloc[]` selects rows and columns by **integer position** (like array indexing).

````python
# Select the first two rows, first two columns
df.iloc[0:2, 0:2]
````

````python
# Select a specific cell: row 1, column 2
df.iloc[1, 2]
````
