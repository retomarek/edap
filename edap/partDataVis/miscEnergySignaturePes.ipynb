# Energy Signature (PES)

```{admonition} Goal
:class: tip
Display a Proposed Energy Signature (PES) with regression lines, following the methodology of Eriksson (2020). The PES relates heating power to outdoor temperature while also considering indoor room temperature.
```

```{admonition} Data Basis
:class: note
The dataset `bldgEngySigProposedData.csv` contains hourly data with columns: `time`, `outsideTemp`, `electricityHeatPump`, `roomTemp` (separator `,`). We build a 4-column DataFrame [timestamp, outside_temp, power, room_temp].
```

## Data Preparation

````python
import pandas as pd

url = "https://raw.githubusercontent.com/retomarek/edap/main/edap/sampleData/bldgEngySigProposedData.csv"
raw = pd.read_csv(url, sep=",")

# Parse timestamp
raw["time"] = pd.to_datetime(raw["time"], utc=True).dt.tz_localize(None)

# Build 4-column DataFrame [timestamp, outside_temp, power, room_temp]
df = raw[["time", "outsideTemp", "electricityHeatPump", "roomTemp"]].copy()
df.columns = ["timestamp", "outside_temp", "power", "room_temp"]
df = df.dropna()

print(f"{len(df)} rows")
df.head()
````

## Visualization

````python
from pyedautils.plots.energy import plot_energy_signature_pes

fig = plot_energy_signature_pes(df, title="Proposed Energy Signature (PES)")
fig.show()
````
