# Calendar Heat Map

```{admonition} Goal
:class: tip
Visualize daily electricity consumption as a calendar heatmap to identify patterns and anomalies over time.
```

```{admonition} Data Basis
:class: note
The dataset `flatElectricity.csv` contains cumulative electricity meter readings at 15-minute intervals for multiple flats. We use the column `FlatA_Ele` and convert cumulative readings to consumption by computing differences.
```

## Data Preparation

````python
import pandas as pd

url = "https://raw.githubusercontent.com/retomarek/edap/main/edap/sampleData/flatElectricity.csv"
raw = pd.read_csv(url, sep=";")

# Parse timestamp and set as index
raw["time"] = pd.to_datetime(raw["time"])
raw = raw.set_index("time")

# Compute 15-min consumption from cumulative meter readings
consumption = raw.diff()

# Select FlatA_Ele and build a clean 2-column DataFrame [timestamp, value]
df = consumption[["FlatA_Ele"]].reset_index()
df.columns = ["timestamp", "value"]
df = df.dropna()

df.head()
````

## Visualization

````python
from pyedautils.plots.heatmaps import plot_heatmap_calendar

fig = plot_heatmap_calendar(df, title="Electricity — Calendar Heat Map", ylab="Electricity [kWh]")
fig.show()
````

::::{dropdown} Customization
The `plot_heatmap_calendar` function accepts a `colorscale` parameter to change the color palette of the heatmap. For example:

```python
fig = plot_heatmap_calendar(df, title="Electricity — Calendar Heat Map", ylab="Electricity [kWh]", colorscale="Viridis")
```
