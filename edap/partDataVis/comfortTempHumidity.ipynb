# Temperature vs. Humidity

```{admonition} Goal
:class: tip
Scatter plot of indoor temperature against relative humidity with comfort zone boundaries. This helps evaluate whether indoor conditions fall within acceptable thermal comfort ranges.
```

```{admonition} Data Basis
:class: note
The dataset `flatTempHum.csv` contains hourly indoor temperature and humidity measurements for multiple flats (columns: `time`, `FlatA_Hum`, `FlatA_Temp`, `FlatB_Hum`, `FlatB_Temp`, …; separator `;`). We use `FlatA_Temp` and `FlatA_Hum` and build a 3-column DataFrame [timestamp, temperature, humidity].
```

## Data Preparation

````python
import pandas as pd

url = "https://raw.githubusercontent.com/retomarek/edap/main/edap/sampleData/flatTempHum.csv"
raw = pd.read_csv(url, sep=";")

# Parse timestamp
raw["time"] = pd.to_datetime(raw["time"])

# Select [timestamp, temperature, humidity] — 3-column DataFrame
df = raw[["time", "FlatA_Temp", "FlatA_Hum"]].copy()
df.columns = ["timestamp", "temperature", "humidity"]
df = df.dropna()

df.head()
````

## Visualization

````python
from pyedautils.plots import plot_comfort_temp_humidity

fig = plot_comfort_temp_humidity(df, title="Indoor Comfort — Temperature vs. Humidity")
fig.show()
````
