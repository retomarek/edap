# Statistical Density

Visualize the probability distribution of a single variable using a density histogram.

```{admonition} Goal
:class: tip
Visualize the probability distribution of indoor temperature for Flat A. A density plot shows the shape of the data distribution, revealing whether temperatures are normally distributed, skewed, or multimodal.
```

```{admonition} Data Basis
:class: note
**Dataset:** `flatTempHum.csv` — Hourly indoor temperature and humidity measurements.  
**Column used:** `FlatA_Temp`  
**Time range:** October 2018 – September 2020
```

## Data Preparation

````python
import pandas as pd

url = "https://raw.githubusercontent.com/retomarek/edap/main/edap/sampleData/flatTempHum.csv"
df = pd.read_csv(url, sep=";")
df["time"] = pd.to_datetime(df["time"])
df = df.set_index("time", drop=True)
df.head()
````

## Visualization

````python
import plotly.express as px

fig = px.histogram(
    df,
    x="FlatA_Temp",
    nbins=50,
    histnorm="probability density",
    title="Temperature Density — Flat A",
    labels={"FlatA_Temp": "Temperature [°C]"},
)
fig.show()
````

```{dropdown} Customization
You can enhance the density visualization:

- Add `marginal="violin"` or `marginal="box"` to `px.histogram()` for a combined view.
- Use `marginal="rug"` to show individual data points along the x-axis.
- Adjust the number of bins with `nbins` to control the smoothness of the histogram.
- For a pure KDE curve, use `plotly.figure_factory.create_distplot()` with `show_hist=False`.
```
