# Mean Profile

```{admonition} Goal
:class: tip
Display the average daily electricity consumption profile with one line per weekday to compare typical usage across the week.
```

```{admonition} Data Basis
:class: note
Electricity consumption data (`ele_meter.csv`) bundled with the `pyedautils` package. 15-minute intervals over ~1 year.
```

## Data Preparation

````python
import pandas as pd
from importlib import resources

data_path = resources.files("pyedautils") / "data" / "ele_meter.csv"
df = pd.read_csv(data_path, sep=";")
df.head()
````

## Visualization

````python
from pyedautils.plots.profiles import plot_daily_profiles

fig = plot_daily_profiles(
    data=df,
    method="mean",
    loc_time_zone="Europe/Zurich",
    title="Electricity - Mean Daily Profiles",
    ylab="Energy Consumption (Wh/15min)",
)
fig.show()
````
