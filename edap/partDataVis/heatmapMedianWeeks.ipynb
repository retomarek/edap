# Median Weeks

```{admonition} Goal
:class: tip
Show the median weekly electricity consumption pattern as a heatmap with hours on one axis and weekdays on the other.
```

```{admonition} Data Basis
:class: note
The dataset `flatElectricity.csv` contains cumulative electricity meter readings at 15-minute intervals. We convert cumulative readings to consumption values for `FlatA_Ele`.
```

## Data Preparation

````python
import pandas as pd

url = "https://raw.githubusercontent.com/retomarek/edap/main/edap/sampleData/flatElectricity.csv"
raw = pd.read_csv(url, sep=";")

# Parse timestamp and set as index
raw.iloc[:, 0] = pd.to_datetime(raw.iloc[:, 0])
raw = raw.set_index(raw.columns[0])

# Compute 15-min consumption from cumulative meter readings
consumption = raw.diff()

# Select FlatA_Ele and build a clean 2-column DataFrame [timestamp, value]
df = consumption[["FlatA_Ele"]].reset_index()
df.columns = ["timestamp", "value"]
df = df.dropna()

df.head()
````

## Visualization

````python
from pyedautils.plots import plot_heatmap_median_weeks

fig = plot_heatmap_median_weeks(df, title="Electricity — Median Weeks", ylab="Electricity [kWh]")
fig.show()
````

```{dropdown} Customization
The `plot_heatmap_median_weeks` function accepts the following optional parameters:

- `colorscale`: Change the color palette (e.g., `"Viridis"`, `"Plasma"`).
- `seasons`: Control how seasons are grouped.

```python
fig = plot_heatmap_median_weeks(df, title="Electricity — Median Weeks", ylab="Electricity [kWh]", colorscale="Plasma", seasons=True)
