# Decomposed Profile

```{admonition} Goal
:class: tip
Display daily electricity consumption profiles with the trend component removed, isolating the intra-day pattern from long-term drift.
```

```{admonition} Data Basis
:class: note
Electricity consumption data (`ele_meter.csv`) bundled with the `pyedautils` package. 15-minute intervals over ~1 year.
```

## Data Preparation

````python
import pandas as pd
from importlib import resources

data_path = resources.files("pyedautils") / "data" / "ele_meter.csv"
df = pd.read_csv(data_path, sep=";")
df.head()
````

## Visualization

````python
from pyedautils.plots.profiles import plot_daily_profiles

fig = plot_daily_profiles(
    data=df,
    method="decomposed",
    loc_time_zone="Europe/Zurich",
    title="Electricity - Decomposed Weekly Pattern",
    ylab="delta Energy Consumption (Wh/15min)",
)
fig.show()
````

```{dropdown} Customization
The `plot_daily_profiles` function with `method="decomposed"` accepts a `k` parameter to control the rolling window size used for trend estimation:

```python
fig = plot_daily_profiles(df, method="decomposed", loc_time_zone="Europe/Zurich", title="Electricity — Decomposed Daily Profiles", ylab="Electricity [kWh]", k=30)
```
