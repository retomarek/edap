# Decomposed Profile

```{admonition} Goal
:class: tip
Display daily electricity consumption profiles with the trend component removed, isolating the intra-day pattern from long-term drift.
```

```{admonition} Data Basis
:class: note
The dataset `flatElectricity.csv` contains cumulative electricity meter readings at 15-minute intervals. We convert cumulative readings to consumption values for `FlatA_Ele`.
```

## Data Preparation

````python
import pandas as pd

url = "https://raw.githubusercontent.com/retomarek/edap/main/edap/sampleData/flatElectricity.csv"
raw = pd.read_csv(url, sep=";")

# Parse timestamp and set as index
raw.iloc[:, 0] = pd.to_datetime(raw.iloc[:, 0])
raw = raw.set_index(raw.columns[0])

# Compute 15-min consumption from cumulative meter readings
consumption = raw.diff()

# Select FlatA_Ele and build a clean 2-column DataFrame [timestamp, value]
df = consumption[["FlatA_Ele"]].reset_index()
df.columns = ["timestamp", "value"]
df = df.dropna()

df.head()
````

## Visualization

````python
from pyedautils.plots import plot_daily_profiles

fig = plot_daily_profiles(df, method="decomposed", loc_time_zone="Europe/Zurich", title="Electricity — Decomposed Daily Profiles", ylab="Electricity [kWh]")
fig.show()
````

```{dropdown} Customization
The `plot_daily_profiles` function with `method="decomposed"` accepts a `k` parameter to control the rolling window size used for trend estimation:

```python
fig = plot_daily_profiles(df, method="decomposed", loc_time_zone="Europe/Zurich", title="Electricity — Decomposed Daily Profiles", ylab="Electricity [kWh]", k=30)
```
