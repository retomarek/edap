# Density Plot by Season

Compare how a variable's distribution shifts across seasons using overlaid density plots.

```{admonition} Goal
:class: tip
Compare the temperature distribution of Flat A across the four seasons. This reveals how indoor temperatures shift between winter and summer and whether the building maintains consistent conditions year-round.
```

```{admonition} Data Basis
:class: note
**Dataset:** `flatTempHum.csv` — Hourly indoor temperature and humidity measurements.  
**Columns used:** `time`, `FlatA_Temp`  
**Time range:** October 2018 – September 2020  
**Library:** `pyedautils.plots.plot_density_seasons` — expects a 2-column DataFrame (timestamp column + value column).
```

## Data Preparation

````python
import pandas as pd

url = "https://raw.githubusercontent.com/retomarek/edap/main/edap/sampleData/flatTempHum.csv"
df = pd.read_csv(url, sep=";", usecols=["time", "FlatA_Temp"])
df["time"] = pd.to_datetime(df["time"])
df.head()
````

## Visualization

````python
from pyedautils.plots import plot_density_seasons

fig = plot_density_seasons(
    df,
    title="Temperature Density by Season",
    xlab="Temperature [°C]",
)
fig.show()
````

```{dropdown} Customization
You can pass a `colors` parameter to `plot_density_seasons()` to override the default season colors. For example:

```python
fig = plot_density_seasons(
    df,
    title="Temperature Density by Season",
    xlab="Temperature [°C]",
    colors={"Winter": "blue", "Spring": "green", "Summer": "red", "Autumn": "orange"},
)
```
