# Overview

```{admonition} Goal
:class: tip
Display a 4x7 grid of daily electricity consumption profiles, organized by season (rows) and weekday (columns), to reveal seasonal and weekly patterns.
```

```{admonition} Data Basis
:class: note
Electricity consumption data (`ele_meter.csv`) bundled with the `pyedautils` package. 15-minute intervals over ~1 year.
```

## Data Preparation

````python
import pandas as pd
from importlib import resources

data_path = resources.files("pyedautils") / "data" / "ele_meter.csv"
df = pd.read_csv(data_path, sep=";")
df.head()
````

## Visualization

````python
from pyedautils.plots.profiles import plot_daily_profiles_overview

fig = plot_daily_profiles_overview(
    data=df,
    title="Electricity Meter - Daily Profiles",
    ylab="Energy [kWh]",
    confidence=95.0,
)
fig.show()
````

```{dropdown} Customization
The `plot_daily_profiles_overview` function accepts the following optional parameters:

- `confidence`: Set the confidence interval band (e.g., `0.95` for 95%).
- `colors`: Customize the color scheme.

```python
fig = plot_daily_profiles_overview(df, title="Electricity — Daily Profiles Overview", ylab="Electricity [kWh]", confidence=0.9)
