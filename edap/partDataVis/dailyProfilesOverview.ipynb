# Daily Profiles Overview

```{admonition} Goal
:class: tip
Display a 4x7 grid of daily electricity consumption profiles, organized by season (rows) and weekday (columns), to reveal seasonal and weekly patterns.
```

```{admonition} Data Basis
:class: note
The dataset `flatElectricity.csv` contains cumulative electricity meter readings at 15-minute intervals. We convert cumulative readings to consumption values for `FlatA_Ele`.
```

## Data Preparation

````python
import pandas as pd

url = "https://raw.githubusercontent.com/retomarek/edap/main/edap/sampleData/flatElectricity.csv"
raw = pd.read_csv(url, sep=";")

# Parse timestamp and set as index
raw.iloc[:, 0] = pd.to_datetime(raw.iloc[:, 0])
raw = raw.set_index(raw.columns[0])

# Compute 15-min consumption from cumulative meter readings
consumption = raw.diff()

# Select FlatA_Ele and build a clean 2-column DataFrame [timestamp, value]
df = consumption[["FlatA_Ele"]].reset_index()
df.columns = ["timestamp", "value"]
df = df.dropna()

df.head()
````

## Visualization

````python
from pyedautils.plots import plot_daily_profiles_overview

fig = plot_daily_profiles_overview(df, title="Electricity — Daily Profiles Overview", ylab="Electricity [kWh]")
fig.show()
````

```{dropdown} Customization
The `plot_daily_profiles_overview` function accepts the following optional parameters:

- `confidence`: Set the confidence interval band (e.g., `0.95` for 95%).
- `colors`: Customize the color scheme.

```python
fig = plot_daily_profiles_overview(df, title="Electricity — Daily Profiles Overview", ylab="Electricity [kWh]", confidence=0.9)
