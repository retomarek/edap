# Comfort — Mollier h-x Diagram

```{admonition} Goal
:class: tip
Display measured indoor conditions on a Mollier h-x (psychrometric) diagram. This visualization shows the relationship between temperature, humidity, and enthalpy of moist air.
```

```{admonition} Data Basis
:class: note
The dataset `flatTempHum.csv` contains hourly indoor temperature and humidity measurements (columns: `time`, `FlatA_Hum`, `FlatA_Temp`, …; separator `;`). We use `FlatA_Hum` (relative humidity in %) and `FlatA_Temp` (temperature in °C) and build a 3-column DataFrame [timestamp, humidity, temperature].
```

## Data Preparation

````python
import pandas as pd

url = "https://raw.githubusercontent.com/retomarek/edap/main/edap/sampleData/flatTempHum.csv"
raw = pd.read_csv(url, sep=";")

# Parse timestamp
raw["time"] = pd.to_datetime(raw["time"])

# Select [timestamp, humidity, temperature] — 3-column DataFrame
df = raw[["time", "FlatA_Hum", "FlatA_Temp"]].copy()
df.columns = ["timestamp", "humidity", "temperature"]
df = df.dropna()

df.head()
````

## Visualization

````python
from pyedautils.plots import plot_mollier_hx
from IPython.display import HTML, display

html = plot_mollier_hx(df, height=700)
display(HTML(html))
````

```{dropdown} Customization
The `plot_mollier_hx` function returns an HTML string containing a D3.js visualization. You can adjust the `height` parameter to change the diagram size.

```python
