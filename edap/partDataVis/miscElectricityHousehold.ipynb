# Electricity Household

```{admonition} Goal
:class: tip
Visualize household electricity consumption patterns over time using a line chart. This provides an overview of how electricity usage varies across the measurement period.
```

```{admonition} Data Basis
:class: note
The dataset `eboBookEleMeter.csv` contains cumulative electricity meter readings at 15-minute intervals (columns: `timestamp`, `eleMeterVal`; separator `;`). We compute consumption by differencing the cumulative values.
```

## Data Preparation

````python
import pandas as pd

url = "https://raw.githubusercontent.com/retomarek/edap/main/edap/sampleData/eboBookEleMeter.csv"
raw = pd.read_csv(url, sep=";")

# Parse timestamp and set as index
raw["timestamp"] = pd.to_datetime(raw["timestamp"])
raw = raw.set_index("timestamp")

# Compute consumption from cumulative meter readings
consumption = raw.diff()

# Resample to hourly sums for a cleaner chart
hourly = consumption.resample("h").sum()
hourly = hourly.dropna()

hourly.head()
````

## Visualization

````python
from pyedautils.plots.stat import plot_timeseries

fig = plot_timeseries(
    hourly,
    columns=["eleMeterVal"],
    title="Household Electricity Consumption",
    ylab="Electricity [kWh/h]",
)
fig.show()
````

```{dropdown} Customization
You can change the resampling frequency to see different time granularities:

```python
# Daily totals
daily = consumption.resample("D").sum()

# Weekly totals
weekly = consumption.resample("W").sum()
