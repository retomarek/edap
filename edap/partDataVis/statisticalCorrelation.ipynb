# Correlation & Scatter

```{admonition} Goal
:class: tip
Explore linear relationships between variables using a correlation matrix heatmap and scatter plots. Correlation analysis reveals which sensors move together, which are independent, and whether relationships are positive or negative.
```

```{admonition} Data Basis
:class: note
**Dataset:** `flatTempHum.csv` — Hourly indoor temperature and humidity measurements from four residential flats (A–D).  
**Columns used:** All temperature and humidity columns  
**Time range:** October 2018 – September 2020
```

## Data Preparation

````python
import pandas as pd

url = "https://raw.githubusercontent.com/retomarek/edap/main/edap/sampleData/flatTempHum.csv"
df = pd.read_csv(url, sep=";")
df["time"] = pd.to_datetime(df["time"])
df = df.set_index("time", drop=True)
df.head()
````

## Correlation Matrix

The correlation matrix shows the Pearson correlation coefficient between every pair of variables. Values range from −1 (perfect negative correlation) to +1 (perfect positive correlation).

````python
from pyedautils.plots.stat import plot_correlation

fig = plot_correlation(df, title="Correlation Matrix — All Sensors")
fig.show()
````

```{dropdown} Interpreting the correlation matrix
- **Dark red (+1)** — variables move together (e.g. temperatures in adjacent flats)
- **Dark blue (−1)** — variables move in opposite directions (e.g. temperature ↑, humidity ↓)
- **White (0)** — no linear relationship
- Temperature–humidity pairs typically show negative correlation because warmer air lowers relative humidity
```

## Scatter Plot

A scatter plot reveals the shape and density of the relationship between two variables. Low opacity makes dense regions visible.

````python
from pyedautils.plots.stat import plot_scatter

fig = plot_scatter(
    df,
    x_column="FlatA_Temp",
    y_column="FlatA_Hum",
    title="Flat A: Temperature vs. Humidity",
    xlab="Temperature [°C]",
    ylab="Humidity [%]",
)
fig.show()
````

````python
fig = plot_scatter(
    df,
    x_column="FlatA_Temp",
    y_column="FlatB_Temp",
    title="Flat A vs. Flat B — Temperature",
    xlab="Flat A [°C]",
    ylab="Flat B [°C]",
)
fig.show()
````

```{dropdown} Customization
- Adjust `opacity` (0.0–1.0) to control how dense regions appear. Lower values are better for large datasets.
- The `color_scale` parameter in `plot_correlation()` accepts any Plotly colorscale (e.g. `"Viridis"`, `"Plasma"`, `"RdBu_r"`).
- Use `text_format=".3f"` for more decimal places in the correlation matrix.
```
