# Overlayed Profiles

```{admonition} Goal
:class: tip
Overlay daily electricity consumption profiles colored by different criteria to compare individual days against each other.
```

```{admonition} Data Basis
:class: note
The dataset `flatElectricity.csv` contains cumulative electricity meter readings at 15-minute intervals. We convert cumulative readings to consumption values for `FlatA_Ele`.
```

## Data Preparation

````python
import pandas as pd

url = "https://raw.githubusercontent.com/retomarek/edap/main/edap/sampleData/flatElectricity.csv"
raw = pd.read_csv(url, sep=";")

# Parse timestamp and set as index
raw.iloc[:, 0] = pd.to_datetime(raw.iloc[:, 0])
raw = raw.set_index(raw.columns[0])

# Compute 15-min consumption from cumulative meter readings
consumption = raw.diff()

# Select FlatA_Ele and build a clean 2-column DataFrame [timestamp, value]
df = consumption[["FlatA_Ele"]].reset_index()
df.columns = ["timestamp", "value"]
df = df.dropna()

df.head()
````

## Visualization

Note: The `method="overlayed"` option returns an HTML string instead of a Plotly Figure.

````python
from pyedautils.plots import plot_daily_profiles
from IPython.display import HTML, display

html = plot_daily_profiles(df, method="overlayed", loc_time_zone="Europe/Zurich", title="Electricity — Overlayed Daily Profiles", ylab="Electricity [kWh]")
display(HTML(html))
````
