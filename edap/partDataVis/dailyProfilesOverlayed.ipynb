# Overlayed Profiles

```{admonition} Goal
:class: tip
Overlay all individual daily electricity consumption profiles. Hovering highlights the entire day-line in orange for comparison.
```

```{admonition} Data Basis
:class: note
Electricity consumption data (`ele_meter.csv`) bundled with the `pyedautils` package. 15-minute intervals over ~1 year.
```

## Data Preparation

````python
import pandas as pd
from importlib import resources

data_path = resources.files("pyedautils") / "data" / "ele_meter.csv"
df = pd.read_csv(data_path, sep=";")
df.head()
````

## Visualization

Note: The `method="overlayed"` option returns an HTML string instead of a Plotly Figure.

````python
from pyedautils.plots.profiles import plot_daily_profiles
from IPython.display import HTML

html = plot_daily_profiles(
    data=df,
    method="overlayed",
    title="Electricity - Overlayed Daily Profiles",
    ylab="Energy Consumption (Wh/15min)",
)
HTML(html)
````
