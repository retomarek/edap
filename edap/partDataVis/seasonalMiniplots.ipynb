# Miniplots

```{admonition} Goal
:class: tip
Show each month as a mini panel to compare energy values across years, making it easy to spot month-by-month trends.
```

```{admonition} Data Basis
:class: note
Monthly energy consumption data (`seasonal_monthly.csv`) bundled with the `pyedautils` package. 10 years of monthly values.
```

## Data Preparation

````python
import pandas as pd
from importlib import resources

data_path = resources.files("pyedautils") / "data" / "seasonal_monthly.csv"
df = pd.read_csv(data_path, sep=";")
df.head()
````

## Visualization

````python
from pyedautils.plots.seasonal import plot_seasonal_miniplots

fig = plot_seasonal_miniplots(
    data=df,
    title="Seasonal Miniplots (10 Years)",
    ylab="Avg Daily Energy (kWh/day)",
)
fig.show()
````
