# Miniplots

```{admonition} Goal
:class: tip
Show each month as a mini panel to compare heating values across years, making it easy to spot month-by-month trends.
```

```{admonition} Data Basis
:class: note
Central heating power measurements (`centralHeating.csv`) resampled to monthly mean values.
```

## Data Preparation

````python
import pandas as pd

url = "https://raw.githubusercontent.com/retomarek/edap/main/edap/sampleData/centralHeating.csv"
df = pd.read_csv(url, sep=";")
df["time"] = pd.to_datetime(df["time"])
df = df.set_index("time")
df = df[["energyHeatingMeter"]]

# Resample to monthly mean
df = df.resample("MS").mean()

df = df.reset_index()
df.columns = ["timestamp", "value"]
df = df.dropna()
df.head()
````

## Visualization

````python
from pyedautils.plots.seasonal import plot_seasonal_miniplots

fig = plot_seasonal_miniplots(df, title="Heating — Seasonal Miniplots", ylab="Heating Power [kW]")
fig.show()
````
