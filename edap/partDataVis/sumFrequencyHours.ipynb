# Hourly Duration Curve

```{admonition} Goal
:class: tip
Display a duration curve (sum frequency) of hourly electricity consumption. The curve ranks hourly totals from highest to lowest, showing the distribution of consumption intensities.
```

```{admonition} Data Basis
:class: note
The dataset `flatElectricity.csv` contains cumulative electricity meter readings at 15-minute intervals (columns: `time`, `FlatA_Ele`, `FlatB_Ele`, `FlatC_Ele`, `FlatD_Ele`; separator `;`). We use `FlatA_Ele`, compute consumption via `diff()`, and resample to hourly sums.
```

## Data Preparation

````python
import pandas as pd

url = "https://raw.githubusercontent.com/retomarek/edap/main/edap/sampleData/flatElectricity.csv"
raw = pd.read_csv(url, sep=";")

# Parse timestamp and set as index
raw.iloc[:, 0] = pd.to_datetime(raw.iloc[:, 0])
raw = raw.set_index(raw.columns[0])

# Compute consumption from cumulative meter readings
consumption = raw.diff()

# Select FlatA_Ele, resample to hourly sums
hourly = consumption[["FlatA_Ele"]].resample("h").sum()

# Build clean 2-column DataFrame [timestamp, value]
df = hourly.reset_index()
df.columns = ["timestamp", "value"]
df = df.dropna()

df.head()
````

## Visualization

````python
from pyedautils.plots import plot_sum_frequency

fig = plot_sum_frequency(df, resolution="hourly", title="Electricity — Hourly Duration Curve", ylab="Electricity [kWh/h]")
fig.show()
````
