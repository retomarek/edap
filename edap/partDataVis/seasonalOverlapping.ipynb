# Overlapping Years

```{admonition} Goal
:class: tip
Overlay multiple years of heating data on the same axis to compare seasonal patterns across years.
```

```{admonition} Data Basis
:class: note
Central heating power measurements (`centralHeating.csv`) resampled to monthly mean values.
```

## Data Preparation

````python
import pandas as pd

url = "https://raw.githubusercontent.com/retomarek/edap/main/edap/sampleData/centralHeating.csv"
df = pd.read_csv(url, sep=";")
df["time"] = pd.to_datetime(df["time"])
df = df.set_index("time")
df = df[["energyHeatingMeter"]]

# Resample to monthly mean
df = df.resample("MS").mean()

df = df.reset_index()
df.columns = ["timestamp", "value"]
df = df.dropna()
df.head()
````

## Visualization

````python
from pyedautils.plots import plot_seasonal_overlapping

fig = plot_seasonal_overlapping(df, title="Heating \u2014 Overlapping Years", ylab="Heating Power [kW]")
fig.show()
````
