# Overlapping Years

```{admonition} Goal
:class: tip
Overlay multiple years of energy data on the same axis to compare seasonal patterns across years.
```

```{admonition} Data Basis
:class: note
Monthly energy consumption data (`seasonal_monthly.csv`) bundled with the `pyedautils` package. 10 years of monthly values.
```

## Data Preparation

````python
import pandas as pd
from importlib import resources

data_path = resources.files("pyedautils") / "data" / "seasonal_monthly.csv"
df = pd.read_csv(data_path, sep=";")
df.head()
````

## Visualization

````python
from pyedautils.plots.seasonal import plot_seasonal_overlapping

fig = plot_seasonal_overlapping(
    data=df,
    title="Seasonal Overlapping (10 Years)",
    ylab="Avg Daily Energy (kWh/day)",
)
fig.show()
````
