# Before / After Comparison

```{admonition} Goal
:class: tip
Compare heating energy consumption before and after an optimization event to evaluate its impact.
```

```{admonition} Data Basis
:class: note
Flat heat and hot water energy measurements (`flatHeatAndHotWater.csv`) resampled to monthly mean values.
```

## Data Preparation

````python
import pandas as pd

url = "https://raw.githubusercontent.com/retomarek/edap/main/edap/sampleData/flatHeatAndHotWater.csv"
df = pd.read_csv(url, sep=";", decimal=",")

# Parse timestamp and select one energy column
df["timestamp"] = pd.to_datetime(df["timestamp"])
df = df.set_index("timestamp")
df = df[["Adr01_energyHeat"]].rename(columns={"Adr01_energyHeat": "value"})

# Resample to monthly mean
df = df.resample("MS").mean()

df = df.reset_index()
df.head()
````

## Visualization

````python
from pyedautils.plots.seasonal import plot_seasonal_before_after

fig = plot_seasonal_before_after(df, date_optimization="2017-09-01", title="Heat & Hot Water — Before/After", ylab="Energy [kWh/day]")
fig.show()
````

```{dropdown} Customization
The `plot_seasonal_before_after` function accepts additional parameters:

- **`date_optimization`** — The date that separates the "before" and "after" periods (format: `"YYYY-MM-DD"`). In this example, `"2017-09-01"` marks the optimization date.
- **`confidence`** — Controls the confidence interval shown around the seasonal means. Adjust to widen or narrow the uncertainty band.
```
