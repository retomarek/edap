# Before / After Comparison

```{admonition} Goal
:class: tip
Compare energy consumption before and after an optimization event to evaluate its impact.
```

```{admonition} Data Basis
:class: note
Monthly energy consumption data (`seasonal_monthly.csv`) bundled with the `pyedautils` package. 10 years of monthly values.
```

## Data Preparation

````python
import pandas as pd
from importlib import resources

data_path = resources.files("pyedautils") / "data" / "seasonal_monthly.csv"
df = pd.read_csv(data_path, sep=";")
df.head()
````

## Visualization

````python
from pyedautils.plots.seasonal import plot_seasonal_before_after

fig = plot_seasonal_before_after(
    data=df,
    date_optimization="2017-09-01",
    title="Before/After Energy Optimization",
    ylab="Avg Daily Energy (kWh/day)",
)
fig.show()
````

::::{dropdown} Customization
The `plot_seasonal_before_after` function accepts additional parameters:

- **`date_optimization`** — The date that separates the "before" and "after" periods (format: `"YYYY-MM-DD"`). In this example, `"2017-09-01"` marks the optimization date.
- **`confidence`** — Controls the confidence interval shown around the seasonal means. Adjust to widen or narrow the uncertainty band.
