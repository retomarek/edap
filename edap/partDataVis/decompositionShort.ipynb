# Short-Term Decomposition

```{admonition} Goal
:class: tip
Decompose electricity meter data over a shorter period to reveal daily consumption patterns and short-term trends.
```

```{admonition} Data Basis
:class: note
Electricity meter readings (`eboBookEleMeter.csv`) at 15-minute intervals, filtered to a 3-month window.
```

## Data Preparation

````python
import pandas as pd

url = "https://raw.githubusercontent.com/retomarek/edap/main/edap/sampleData/eboBookEleMeter.csv"
df = pd.read_csv(url, sep=";")
df["timestamp"] = pd.to_datetime(df["timestamp"])
df = df.set_index("timestamp")

# Meter is cumulative → consumption = diff
df["value"] = df["eleMeterVal"].diff()
df = df.dropna()
df = df[df["value"] >= 0]

# Filter to 3 months of data
start = df.index.min()
end = start + pd.DateOffset(months=3)
df = df.loc[start:end]

df = df[["value"]].reset_index()
df.columns = ["timestamp", "value"]
df.head()
````

## Visualization

````python
from pyedautils.plots.stat import plot_decomposition

fig = plot_decomposition(df, title="Electricity Consumption — Short-term Decomposition", ylab="Consumption [Wh/15min]")
fig.show()
````
