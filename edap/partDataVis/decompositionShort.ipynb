# Short-term Decomposition

```{admonition} Goal
:class: tip
Decompose outdoor temperature over a shorter period to reveal daily patterns and short-term trends.
```

```{admonition} Data Basis
:class: note
Central outside temperature measurements (`centralOutsideTemp.csv`) filtered to a 3-month window.
```

## Data Preparation

````python
import pandas as pd

url = "https://raw.githubusercontent.com/retomarek/edap/main/edap/sampleData/centralOutsideTemp.csv"
df = pd.read_csv(url, sep=";")
df["time"] = pd.to_datetime(df["time"])
df = df.set_index("time")
df = df[["value"]]

# Filter to 3 months of data
start = df.index.min()
end = start + pd.DateOffset(months=3)
df = df.loc[start:end]

df = df.reset_index()
df.head()
````

## Visualization

````python
from pyedautils.plots import plot_decomposition

fig = plot_decomposition(df, title="Outside Temperature \u2014 Short-term Decomposition", ylab="Temperature [\u00b0C]")
fig.show()
````
