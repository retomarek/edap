# Short-Term Decomposition

```{admonition} Goal
:class: tip
Decompose high-resolution power data over a short period to reveal daily consumption patterns and short-term trends.
```

```{admonition} Data Basis
:class: note
Power measurements at 15-minute intervals (`decomposition_short.csv`) bundled with the `pyedautils` package. 480 data points covering 5 days.
```

## Data Preparation

````python
import pandas as pd
from importlib import resources

data_path = resources.files("pyedautils") / "data" / "decomposition_short.csv"
df = pd.read_csv(data_path, sep=";")
df.head()
````

## Visualization

````python
from pyedautils.plots.stat import plot_decomposition

fig = plot_decomposition(
    data=df,
    period=96,
    s_window=193,
    title="Time Series Decomposition over 5 Days",
    ylab="Power [kW]",
)
fig.show()
````
